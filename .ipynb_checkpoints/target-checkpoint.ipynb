```python
import random
import numpy as np
SEED = 42

# 1. Python built-in random
random.seed(SEED)

# 2. NumPy
np.random.seed(SEED)

import pandas as pd
import re
import os
import warnings
warnings.simplefilter(action='ignore', category=Warning)
pd.set_option("display.max_colwidth", None)  # Don't truncate column values
pd.set_option("display.max_rows", None)  # Don't truncate column values

df_train = pd.read_excel('./RunStatistic.xlsx', skiprows=8)
df_viskos = pd.read_csv('viskos_means.csv', delimiter=";")
df_abrasion = pd.read_csv('abrasion.csv', delimiter=";")

df = pd.DataFrame(columns= ["Run_ID","Experiment",
                 
                 "Basis", "Matt_Beads", "Matt_Silica", "Wachs", "Verdicker",
                 
                 # time from Dispsene until mix & time from Mix to drawdown
                 "Time_To_Mix", "Time_To_Dispense", 
                 
                 # thickness
                "thickness", # README Lesen!!! offset, nicht alle Werte brauchbar
                 
                 # Gloss / Haze
                "gloss_60", "gloss_85", "haze",
                 
                 #Spektrum
                "CIELAB_10deg_L", "CIELAB_10deg_a", "CIELAB_10deg_b", # README Lesen!!! L modellieren, a&b korrelaion prüfen
                ])
i=0
print(df_train["Run ID"].unique())
for run_id in df_train["Run ID"].unique():
    df_temp_run = df_train[df_train["Run ID"] == run_id]
    for exp_id in df_temp_run["Experiment"].unique():
        #if run_id == 864:
        #    display(df_temp_run.head(20))
        i += 1
        df_temp_exp = df_temp_run[df_temp_run["Experiment"] == exp_id]
        df_temp = df_temp_exp
        exp_id = exp_id.split(" ")[1]
        df_temp["Experiment"] == exp_id

        # extract dispensed values
        value1 = df_temp[df_temp['Dispense Material'] == 'I2Dach_Hesse_Basismodul']['Dispensed Amount'].iloc[0]
        value2 = df_temp[df_temp['Dispense Material'] == 'I2Dach_Hesse_Mattmodul Beads HM 9-007']['Dispensed Amount'].iloc[0]
        value3 = df_temp[df_temp['Dispense Material'] == 'I2Dach_Hesse_Mattmodul Silica HM 9-008']['Dispensed Amount'].iloc[0]
        value4 = df_temp[df_temp['Dispense Material'] == 'I2Dach_Hesse_Wachsmodul HM 7-004']['Dispensed Amount'].iloc[0]
        value5 = df_temp[df_temp['Dispense Material'] == 'I2Dach_Hesse_Verdickermodul HZ 3-88']['Dispensed Amount'].iloc[0]

        # Calculate the time until mix
        timestamp_dispense = pd.to_datetime(df_temp[df_temp['Node'] == '3580: Dispense Liquid - M']['Started (UTC)'].iloc[0])
        timestamp_mix = pd.to_datetime(df_temp[df_temp['Node'] == '3583: Mix']['Started (UTC)'].iloc[0])
        time_to_mix = timestamp_mix - timestamp_dispense
        
        # Calculate the time from mix to dispense
        timestamp_drawdown = pd.to_datetime(df_temp[df_temp['Node'] == '3585: Draw Down']['Started (UTC)'].iloc[0])
        time_to_drawdown = timestamp_drawdown - timestamp_mix

        # get the characterization results
        number = re.findall(r'\(#(\d+)\)', df_temp['Experiment'].iloc[0])[0] if re.search(r'\(#(\d+)\)', df_temp['Experiment'].iloc[0]) else None
        contents = os.listdir('measurements/')
        matching_item = next((item for item in contents if number in item), None)
        contents = os.listdir('measurements/'+matching_item)
        
        ### thickness | # Messpunkte 3+4 | GZ nur Zeile 4 (dritter Messwert), Offset beachten (10 abziehen)
        df_thickness = pd.read_csv('measurements/'+matching_item+"/"+next((item for item in contents if "minitest" in item), None), skiprows=0)
        thickness = df_thickness.iloc[2][0] - 10

        ### Gloss / Haze | Messpunkt 2 | GZ: Gloss60, 85, Haze, Dritte Zeile (Zweiter Messwert)
        df_haze_gloss = pd.read_csv('measurements/'+matching_item+"/"+next((item for item in contents if "haze" in item), None), skiprows=0).iloc[:, 1:]
        gloss_60, gloss_85, haze = df_haze_gloss.iloc[1][1], df_haze_gloss.iloc[1][2], df_haze_gloss.iloc[1][3] 

        ### Spectrum | Messpunkte 2+3 | GZ: nur Zeile 3 nehmen (zweiter Messwert)
        df_xrite = pd.read_csv('measurements/'+matching_item+"/"+next((item for item in contents if "xrite" in item), None), skiprows=0).iloc[:, 1:]
        spectrum_1, spectrum_2, spectrum_3 = df_xrite["CIELAB_10deg_L"].iloc[1], df_xrite["CIELAB_10deg_a"].iloc[1], df_xrite["CIELAB_10deg_b"].iloc[1]     
        
        new_row = pd.DataFrame({"Run_ID": run_id,"Experiment": exp_id,
                                "Basis": value1, 
                                "Matt_Beads": value2, 
                                "Matt_Silica": value3, 
                                "Wachs": value4, 
                                "Verdicker": value5,

                                # time from Dispsene until mix & time from Mix to drawdown
                                "Time_To_Mix": time_to_mix.total_seconds(), "Time_To_Dispense": time_to_drawdown.total_seconds(),

                                 # thickness
                                "thickness": thickness,

                                # Haze
                                "gloss_60": gloss_60, "gloss_85": gloss_85, "haze": haze,

                                # Spektrum
                                "CIELAB_10deg_L": spectrum_1 - 4.36, "CIELAB_10deg_a": spectrum_2, "CIELAB_10deg_b": spectrum_3
                                }, index=[0])
        df = pd.concat([df, new_row], ignore_index=True)
df = pd.concat([df, df_viskos], axis=1)
print(len(df))
df = df.rename(columns={'Run_ID': 'run_id', 'Experiment': 'experiment_id'})
df_abrasion = df_abrasion.rename(columns={'RUN_ID': 'run_id', 'EXPERIMENT_ID': 'experiment_id'})

df['run_id'] = df['run_id'].astype(int)
df['experiment_id'] = df['experiment_id'].astype(int)
df_abrasion['run_id'] = df_abrasion['run_id'].astype(int)
df_abrasion['Experiment_ID'] = df_abrasion['Experiment_ID'].astype(int)
df_abrasion = df_abrasion.rename(columns={"Experiment_ID": "experiment_id"})
df = pd.merge(df, df_abrasion, on=['run_id', 'experiment_id'], how='inner')
df
```

```python
import numpy as np
import pandas as pd
import random
import warnings
from itertools import product

from sklearn.multioutput import MultiOutputRegressor
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import (
    RBF, Matern, WhiteKernel, ExpSineSquared, RationalQuadratic
)
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import (
    mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
)

# === Suppress warnings (including GPR convergence warnings) ===
warnings.filterwarnings("ignore")

# === Feature and target selection ===
selected_train_columns = [
    'Basis', 'Matt_Beads', 'Matt_Silica', 'Wachs', 'Verdicker',
    'Time_To_Mix', 'Time_To_Dispense', 'thickness',
    'haze', 'CIELAB_10deg_L', 'CIELAB_10deg_a', 'CIELAB_10deg_b'
]

selected_target_columns = ['gloss_85', 'Viskos2']

# === Extract features and targets ===
X = np.array(df[selected_train_columns])
y = np.array(df[selected_target_columns])

# === Feature scaling ===
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()
X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)

# === Train-test split ===
X_train_scaled, X_test_scaled, y_train_scaled, y_test_scaled = train_test_split(
    X_scaled, y_scaled, test_size=0.2, random_state=SEED, shuffle=True
)

print("Train/Test Shapes:")
print("X_train:", X_train_scaled.shape, "| y_train:", y_train_scaled.shape)
print("X_test :", X_test_scaled.shape, "| y_test :", y_test_scaled.shape)

# === Kernel search space setup ===
ls = 1e-2
nu_values = [1.5, 0.5]
length_scale_values = [0.5 * ls, 2.0 * ls]
noise_level_values = [0.01, 0.1]
hyperparameter_combinations = list(product(nu_values, length_scale_values, noise_level_values))

# === Define kernel list ===
kernels = [
    1.0 * RBF(),
    1.0 * Matern(),
    1.0 * WhiteKernel(),
    1.0 * ExpSineSquared(),
    1.0 * RationalQuadratic(),
    1.0 * (RBF() + WhiteKernel()),
    1.0 * (Matern() + WhiteKernel()),
    1.0 * (ExpSineSquared() + WhiteKernel()),
    1.0 * (RationalQuadratic() + WhiteKernel())
]

# === Add composite kernels with different hyperparameters ===
for nu, length_scale, noise_level in hyperparameter_combinations:
    kernels.append(
        1.0 * Matern(nu=nu) + WhiteKernel(noise_level=noise_level) + RBF(length_scale=length_scale)
    )

# === Grid search parameters ===
param_grid = {
    'estimator__kernel': kernels,
    'estimator__n_restarts_optimizer': [5]
}

# === Base and wrapped model ===
base_gpr = GaussianProcessRegressor(random_state=SEED, normalize_y=True)
multi_gpr = MultiOutputRegressor(base_gpr)

# === GridSearchCV ===
gpr_grid_search = GridSearchCV(
    multi_gpr,
    param_grid,
    cv=3,
    n_jobs=-1,
    scoring='neg_mean_squared_error'
)

print("Starting grid search...")
gpr_grid_search.fit(X_train_scaled, y_train_scaled)

# === Best model and prediction ===
best_model = gpr_grid_search.best_estimator_
y_pred_scaled = best_model.predict(X_test_scaled)

# === Evaluation (scaled space) ===
mse = mean_squared_error(y_test_scaled, y_pred_scaled)
mae = mean_absolute_error(y_test_scaled, y_pred_scaled)
mape = mean_absolute_percentage_error(y_test_scaled, y_pred_scaled)

print("\n✅ Best GPR Hyperparameters:")
print(gpr_grid_search.best_params_)

print("\n📊 Evaluation Metrics (scaled space):")
print(f"MSE  : {mse:.4f}")
print(f"MAE  : {mae:.4f}")
print(f"MAPE : {mape:.4f}")
```
